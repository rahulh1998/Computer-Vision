# Drawing shapes in Images

```python
im1 = np.zeros((512,512,3),dtype = np.int16)
plt.imshow(im1)
```

```python
# Drawing a rectangle requires top left & bottom right coordinates
cv2.rectangle(im1,(100,100),(200,200),color = (0,210,120),thickness=3)
plt.imshow(im1)
```

```python
#thickness = -1 fills the shape
cv2.circle(im1,center = (300,300),radius = 50,color = (0,120,210),thickness = -1)
plt.imshow(im1)
```

# Text on images


```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(im1,'Hey',org = (200,400),fontFace = font,fontScale = 4,color = (255,255,255),thickness = 3)
plt.imshow(im1)
```

## Mouse Drag 


```python
import cv2 
import numpy as np

ix,iy = -1,-1
drawing = False

def draw_mouse(event,x,y,flags,param):
    global ix,iy,drawing
    
    # Mouse click 
    if event == cv2.EVENT_LBUTTONDOWN:
        # assigning the initial coordinates to ix & iy
        ix, iy = x,y
        drawing = True
    
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(image,(ix,iy),(x,y), color = (150,150,0), thickness = -1)
    
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(image,(ix,iy),(x,y), color = (150,150,0), thickness = -1)

image = np.zeros((1000,1000,3), dtype = np.uint8)
        
cv2.namedWindow(winname = 'window2')

cv2.setMouseCallback('window2',draw_mouse)

while True:
    cv2.imshow("window2",image)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
