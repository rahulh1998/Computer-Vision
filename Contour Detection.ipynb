```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Display an image 

def display_gray(img,b=True):
    plt.figure(dpi = 100)
    if b:      
        plt.imshow(img,cmap ='gray')
        plt.show()
    else:
        color = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        plt.imshow(color)
        plt.show()
```

```python
# Read an image

image = cv2.imread('DATA/coins_on_white.jpg')
#plt.imshow(image,cmap='gray')
display_gray(image)
```

```python
# Converting to grayscale

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
display_gray(gray)
```

```python
# Reducing noise by applying blur

blur = cv2.GaussianBlur(image,ksize = (9,9),sigmaX = 0)
#blur = cv2.bilateralFilter(gray,19,100,100)
display_gray(blur)
```

```python
# edge detection

edges = cv2.Canny(blur,50,150)
display_gray(edges)
```

```python
# Finding contours

contours,hierarchy = cv2.findContours(edges,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
image_c = image.copy()
cv2.drawContours(image_c,contours,-1, (0, 255, 0), 3)
display_gray(image_c,False)
```
